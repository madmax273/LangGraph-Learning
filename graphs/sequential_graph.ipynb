```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
```

```python
class State(TypedDict):
    weight: float
    height: float
    BMI: float
```

```python
graph = StateGraph(State)
```

```python
def bmi_calculator(state: State) -> State:
     state["BMI"]= state["weight"] / (state["height"] / 100) ** 2
     return state
```

```python
graph.add_node("bmi_calculator", bmi_calculator)
```

```python

graph.add_edge(START, "bmi_calculator")
graph.add_edge("bmi_calculator", END)
workflow = graph.compile()
output=workflow.invoke({"height": 1.75, "weight": 70})
print(output["BMI"])
```
